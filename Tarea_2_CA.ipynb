# Análisis de emociones con DeepFace

```python
from deepface import DeepFace

def process_image_df(file_name):
  return DeepFace.analyze(file_name)

# Loop through images
for i in arr_images:
  result = process_image_df(image_path + i)

  pprint(result["emotion"])
  print(f'dominant_emotion: { result["dominant_emotion"]}')

  result_image = mpimg.imread(image_path + i)
  imgplot = plt.imshow(result_image)
  
  # Display Output Image
  plt.show()

  print ('\n')
```

# Análisis de emociones con FER

```python
import cv2
from fer import FER

def process_image_fer(file_name):
  
  # Input Image
  input_image = cv2.imread(file_name)
  emotion_detector = FER()
  
  # Save output in result variable
  result = emotion_detector.detect_emotions(input_image)

  # Painting the rectangle
  bounding_box = result[0]["box"]
  emotions = result[0]["emotions"]
  cv2.rectangle(input_image,(bounding_box[0], bounding_box[1]),(bounding_box[0] + bounding_box[2], bounding_box[1] + bounding_box[3]), (0, 155, 255), 2,)

  emotion_name, score = emotion_detector.top_emotion(input_image)
  for index, (emotion_name, score) in enumerate(emotions.items()):
    color = (255, 255, 255) if score < 0.01 else (255, 255, 0)
    emotion_score = "{}: {}".format(emotion_name, "{:.2f}".format(score))

    cv2.putText(input_image,emotion_score, (bounding_box[0], bounding_box[1] + bounding_box[3] + 30 + index * 15), cv2.FONT_HERSHEY_SIMPLEX,0.5,color,1,cv2.LINE_AA,)

  # Save the result in new image file
  new_file_name = str.replace (file_name, ".", "_box.")
  cv2.imwrite(new_file_name, input_image)

  return new_file_name

# Loop through images
for i in arr_images:
  # Read image file using matplotlib's image module
  result_image = mpimg.imread(process_image_fer(image_path + i))
  imgplot = plt.imshow(result_image)
  
  # Display Output Image
  plt.show()
```

# Análisis emociones en Video con FER

```python
from fer import Video
from fer import FER

# https://www.youtube.com/watch?v=B0ouAnmsO1Y
filename = 'The secrets to decoding facial expressions'
video_filename = video_path + filename + ".mp4"

video = Video(video_filename)

# Analyze video, displaying the output
detector = FER(mtcnn=True)
raw_data = video.analyze(detector, display=False, zip_images=False, frequency=50)
```

```python
df = video.to_pandas(raw_data)
df = video.get_first_face(df)
df = video.get_emotions(df)

# Plot emotions
df.plot()
plt.show()
```
